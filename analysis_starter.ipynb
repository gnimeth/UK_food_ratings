## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our new database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result

number = establishments.count_documents(query)

print(f'The number of establishments with a hygiene score of 20 is: {number}')
```

```python
# Find the first document that matches the query
first_document = establishments.find_one(query)

# Use pprint to display the document in a more human-readable format
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(hygiene_20_df))

# Display the first 10 rows of the DataFrame
hygiene_20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
rating_value =  {'LocalAuthorityName': {'$regex':'London'},
                 'RatingValue': {'$gte': '4'}}

# Use count_documents to display the number of documents in the result

rating_value_result = establishments.count_documents(rating_value)

print(f'The number of establishments with a rating value greater than or equal to four: {rating_value_result}')
```

```python
# Find the first document that matches the query
first_one = establishments.find_one(rating_value)

# Use pprint to display the document in a more human-readable format
pprint(first_one)
```

```python
# Convert the result to a Pandas DataFrame
rating_greater_4_df= pd.DataFrame(establishments.find(rating_value))

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(rating_greater_4_df))

# Display the first 10 rows of the DataFrame
rating_greater_4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 1.0
latitude = 51.49014200
longitude = 0.08384000

query = {'$match': {'RatingValue': "5",
                    'geocode.longitude': {'$gte': longitude-degree_search, '$lte': longitude+degree_search},
                    'geocode.latitude': {'$gte': latitude-degree_search, '$lte': latitude+degree_search}}}
sort =  {'$sort': {'scores.Hygiene': 1}}
limit = {'$limit': 5}

# Print the results
results = list(establishments.aggregate([query, sort, limit]))

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
near_penang = pd.DataFrame(results)

near_penang
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 

# 1. Matches establishments with a hygiene score of 0

match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# 2. Groups the matches by Local Authority

group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]
```

```python
# Run the pipeline through the aggregate method, cast the results as a list, and save the results to a variable
hygiene_0 = list(establishments.aggregate(pipeline))
```

```python
# Print the number of rows in the result
print("Number of rows in result: ", len(hygiene_0))
```

```python
# Print the first 10 results
pprint(hygiene_0[0:10])
```

```python
# Convert mongo result to Pandas DataFrame
hygiene_0_df = pd.DataFrame(hygiene_0)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(hygiene_0_df))

# Display the first 10 rows of the DataFrame
hygiene_0_df.head(10)
```
